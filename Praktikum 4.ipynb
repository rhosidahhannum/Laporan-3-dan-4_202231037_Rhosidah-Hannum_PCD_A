### Ekstraksi Fitur

```python
# Mengimpor pustaka OpenCV untuk pengolahan gambar dan visi komputer
import cv2

# Mengimpor pustaka Matplotlib untuk plotting gambar dan visualisasi
import matplotlib.pyplot as plt

# Mengimpor pustaka NumPy untuk operasi numerik dan manipulasi array
import numpy as np

# Mengimpor pustaka skimage untuk pengolahan gambar lanjutan
import skimage

# Mengimpor modul graycomatrix dan graycoprops dari skimage.feature
# Modul ini digunakan untuk menghitung matriks korelasi abu-abu (GLCM)
from skimage.feature import graycomatrix, graycoprops
```

```python
# Mengimpor gambar astronot dari dataset gambar yang disediakan oleh scikit-image
img = skimage.data.astronaut()

# Mengubah mode warna gambar dari RGB (Red-Green-Blue) ke HSV (Hue-Saturation-Value)
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

# Membuat subplot dengan 1 baris dan 2 kolom, ukuran figur 10x10 inci
fig, axs = plt.subplots(1, 2, figsize=(10, 10))

# Meratakan sumbu dari subplot ke dalam array ax
ax = axs.ravel()

# Menampilkan gambar asli dalam mode warna RGB di subplot pertama (ax[0])
ax[0].imshow(img)
ax[0].set_title("RGB")  # Menetapkan judul subplot pertama

# Menampilkan gambar yang telah dikonversi ke mode warna HSV di subplot kedua (ax[1])
# dengan menggunakan colormap "hsv" untuk memvisualisasikan mode warna HSV
ax[1].imshow(img_hsv, cmap="hsv")
ax[1].set_title("HSV")  # Menetapkan judul subplot kedua

# Menampilkan plot dengan kedua subplot yang telah ditetapkan gambar dan judulnya
plt.show()
```

```python
# Menghitung nilai rata-rata dan deviasi standar dari nilai piksel dalam gambar HSV
mean = np.mean(img_hsv.ravel())  # Menghitung nilai rata-rata
std = np.std(img_hsv.ravel())    # Menghitung deviasi standar

# Menampilkan nilai rata-rata dan deviasi standar
print(mean, std)
```

```python
# Mengambil saluran Hue (H) dari gambar HSV
hue = img_hsv[:,:,0]
```

```python
# Menghitung Gray-Level Co-occurrence Matrix (GLCM) dari saluran Hue (H) dalam gambar
# dengan jarak 1 piksel, sudut 0 derajat, 256 level, simetris, dan dinormalisasi
glcm = graycomatrix(hue, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
```

```python
# Menghitung propertis tekstur berdasarkan Gray-Level Co-occurrence Matrix (GLCM)
# yang sudah dihitung sebelumnya dari saluran Hue (H) dalam gambar

# Menghitung nilai contrast dari GLCM
contrast = graycoprops(glcm, 'contrast')[0, 0]

# Menghitung nilai dissimilarity (ketidaksamaan) dari GLCM
dissimilarity = graycoprops(glcm, 'dissimilarity')[0, 0]

# Menghitung nilai homogeneity (homogenitas) dari GLCM
homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]

# Menghitung nilai energy (energi) dari GLCM
energy = graycoprops(glcm, 'energy')[0, 0]

# Menghitung nilai correlation (korelasi) dari GLCM
correlation = graycoprops(glcm, 'correlation')[0, 0]
```

```python
# Mencetak nilai propertis tekstur yang sudah dihitung sebelumnya dari GLCM

# Mencetak nilai contrast yang sudah dihitung sebelumnya
print(f'contrast : {contrast}')

# Mencetak nilai dissimilarity (ketidaksamaan) yang sudah dihitung sebelumnya
print(f'dissimilarity : {dissimilarity}')

# Mencetak nilai homogeneity (homogenitas) yang sudah dihitung sebelumnya
print(f'homogeneity : {homogeneity}')

# Mencetak nilai energy (energi) yang sudah dihitung sebelumnya
print(f'energy : {energy}')

# Mencetak nilai correlation (korelasi) yang sudah dihitung sebelumnya
print(f'correlation : {correlation}')
```
